```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
data_1=pd.read_csv('G:/Data Science/My notes/365 DATA SCIENCE/Project/minutes_watched_2021_paid_0.csv')
data_2=pd.read_csv('G:/Data Science/My notes/365 DATA SCIENCE/Project/minutes_watched_2021_paid_1.csv')
data_3=pd.read_csv('G:/Data Science/My notes/365 DATA SCIENCE/Project/minutes_watched_2022_paid_0.csv')
data_4=pd.read_csv('G:/Data Science/My notes/365 DATA SCIENCE/Project/minutes_watched_2022_paid_1.csv')
```

```python
fig,axes = plt.subplots(2,2,figsize=(15,10))

sns.kdeplot(data_1['minutes_watched'],ax=axes[0,0]).set(title='Free-Plan Students for Q2 2021')
sns.kdeplot(data_3['minutes_watched'],ax=axes[0,1]).set(title='Free-Plan Students for Q2 2022')
sns.kdeplot(data_2['minutes_watched'],ax=axes[1,0]).set(title='Paid Students for Q2 2021')                                                       
sns.kdeplot(data_4['minutes_watched'],ax=axes[1,1]).set(title='Paid Students for Q2 2022')                                                            
                                                        
fig.tight_layout(pad = 3.0)

plt.show()
```

### Removing The Outliers

```python
q1=data_1['minutes_watched'].quantile(0.99)
q2=data_2['minutes_watched'].quantile(0.99)
q3=data_3['minutes_watched'].quantile(0.99)
q4=data_4['minutes_watched'].quantile(0.99)
```

```python
data_no_outliers1=data_1[data_1['minutes_watched']<q1]
data_no_outliers2=data_2[data_2['minutes_watched']<q2]
data_no_outliers3=data_3[data_3['minutes_watched']<q3]
data_no_outliers4=data_4[data_4['minutes_watched']<q4]
```

```python
fig,axes = plt.subplots(2,2,figsize=(15,10))

sns.kdeplot(data_no_outliers1['minutes_watched'],ax=axes[0,0]).set(title='Free-Plan Students for Q2 2021(no_outliers)')
sns.kdeplot(data_no_outliers3['minutes_watched'],ax=axes[0,1]).set(title='Free-Plan Students for Q2 2022(no_outliers)')
sns.kdeplot(data_no_outliers2['minutes_watched'],ax=axes[1,0]).set(title='Paid Students for Q2 2021(no_outliers)')                                                       
sns.kdeplot(data_no_outliers4['minutes_watched'],ax=axes[1,1]).set(title='Paid Students for Q2 2022(no_outliers)')                                                            
                                                        
fig.tight_layout(pad = 3.0)

plt.show()
```

### Saving the data as a CSV file.

```python
data_no_outliers1.to_csv('minutes_watched_2021_paid_0_no_outliers.csv',index=False)
data_no_outliers2.to_csv('minutes_watched_2022_paid_0_no_outliers.csv',index=False)
data_no_outliers3.to_csv('minutes_watched_2021_paid_1_no_outliers.csv',index=False)
data_no_outliers4.to_csv('minutes_watched_2022_paid_1_no_outliers.csv',index=False)
```
